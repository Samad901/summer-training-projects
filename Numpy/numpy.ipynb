```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as pit

img = Image.open("Images 2.jpg")

img = np.array(img)

img[:, :, 0] = np.clip(img[:, :,0] + 0, 0, 55)
pit.imshow(img)
# pit.axis("off")
pit.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("Images 2.jpg")
img = np.array(img)

h, w, c = img.shape
mid_h = h // 2
mid_w = w // 2

img[:mid_h, :mid_w, 0] = np.clip(img[:mid_h, :mid_w, 0] + 60, 0, 255)


img[:mid_h, mid_w:, 1] = np.clip(img[:mid_h, mid_w:, 1] + 60, 0, 255)


img[mid_h:, :mid_w, 2] = np.clip(img[mid_h:, :mid_w, 2] + 60, 0, 255)


img[mid_h:, mid_w:, :] = np.clip(img[mid_h:, mid_w:, :].astype(int) - 40, 0, 255)

plt.imshow(img)
plt.show()
```
