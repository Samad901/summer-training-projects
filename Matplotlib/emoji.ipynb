```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Wedge


fig, ax = plt.subplots(figsize=(5, 5))
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.axis('off')


face = Circle((0, 0), 1, color='yellow', ec='black', linewidth=2)
ax.add_patch(face)

left_eye = Circle((-0.35, 0.35), 0.15, color='black')
ax.add_patch(left_eye)

right_eye = Circle((0.35, 0.35), 0.15, color='black')
ax.add_patch(right_eye)

sad_mouth = Wedge((0, -0.1), 0.5, 180, 360, color='black',linewidth=3, fill=False)
ax.add_patch(sad_mouth)


plt.show()
```
